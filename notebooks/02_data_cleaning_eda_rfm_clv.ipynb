```python
final_df["line_revenue"]=final_df["unit_price"]*final_df["quantity"]
final_df[["unit_price","quantity","line_revenue"]].head()
```

```python
final_df["line_cost"]=final_df["cost_price"]*final_df["quantity"]
final_df[["cost_price","quantity","line_cost"]].head()
```

```python
final_df["discount_value"]=(final_df["discount_percent"]/100)*final_df["unit_price"]*final_df["quantity"]
final_df[["unit_price","quantity","discount_percent","discount_value"]]
```

```python
final_df["profit_margin_percent"]=np.where(final_df["line_revenue"]>0,(final_df["profit"]/final_df["line_revenue"])*100,0)
final_df[["line_revenue","profit","profit_margin_percent"]].round(2) 
final_df[["line_revenue","profit","profit_margin_percent"]].head(10).round(2)
```

```python
final_df["order_year"]=final_df["order_date"].dt.year
final_df["order_month"]=final_df["order_date"].dt.month
final_df["order_quarter"]=final_df["order_date"].dt.quarter

final_df[["order_date","order_year","order_month","order_quarter"]].head()
```

```python
# lets find mponthly revenue 
monthly_revenue=(final_df.groupby(["order_year","order_month"])["line_revenue"].sum().reset_index().sort_values(["order_year","order_month"]))
print(monthly_revenue)
monthly_revenue.head(10)
```

```python
# Lets find some KPI's which includes some tops numbers  using pandas 

#TOTAL REVENUE
total_revenue=final_df["line_revenue"].sum().round(2)

# TOTAL PROFIT
total_profit=final_df["profit"].sum().round(2)


#TOTAL DISCOUNT GIVEN
total_discount=final_df["discount_value"].sum().round(2)


# TOTAL UNIQUE ORDERS
total_orders=final_df["order_id"].nunique()


# TOTAL UNIQUE CUSTOMERS
total_customers=final_df["customer_id"].nunique()


#AVERAGE REVEUE PER ORDER
avg_revenue_per_order=total_revenue/total_orders

print("Total Revenue:", total_revenue)
print("Total Profit:", total_profit)
print("Total Discount Given:", total_discount)
print("Total Orders:", total_orders)
print("Total Customers:", total_customers)
print("Average Revenue Per Order:", avg_revenue_per_order)
```

```python
# Segement Wise Discount Impact

segment_discount_impact=final_df.groupby("segment").agg(avg_discount_percent=("discount_percent","mean"),avg_profit_margin=("profit_margin_percent","mean"),total_revenue=("line_revenue","sum")).reset_index().sort_values("total_revenue",ascending=False).round(2)
segment_discount_impact.head()
```

```python
# RFM Analysis (Recency, Frequency, Monetary)

rfm=final_df.groupby("customer_id").agg(last_order_date=("order_date","max"),total_orders=("order_id","nunique"),total_revenue=("line_revenue","sum")).reset_index().sort_values("total_revenue",ascending=False)
rfm
```

```python
# RFM Scoring


max_date=rfm["last_order_date"].max()
rfm["recency_days"]=(max_date-rfm["last_order_date"]).dt.days
rfm.head()

# Recency Score
rfm["R_Score"]=pd.qcut(rfm["recency_days"],5,labels=[5,4,3,2,1]).astype(int)

# Frequency Score
rfm["F_Score"]=pd.qcut(rfm["total_orders"],5,labels=[1,2,3,4,5]).astype(int)

# Monetary Score
rfm["M_Score"]=pd.qcut(rfm["total_revenue"],5,labels=[1,2,3,4,5]).astype(int)

rfm.head()
```

```python
rfm["RFM_Score"]=rfm["R_Score"]+rfm["F_Score"]+rfm["M_Score"]
rfm["Segment"]=np.select(
    [(rfm["RFM_Score"]>=13),
     (rfm["RFM_Score"]>=10)  & (rfm["R_Score"]>=4),
     (rfm["RFM_Score"]>=10) &(rfm["R_Score"]<=3),
     (rfm["RFM_Score"] >= 7),                                                            
     (rfm["RFM_Score"] >= 5),                                                             
     (rfm["RFM_Score"] < 5)
    ],
    [
        "Champions",               # High R, High F, High M
        "Loyal Customers",
        "Potential Loyalists",
        "At Risk",
        "Hibernating",
        "Lost / Churn"
    ],
    default="Others"
)
rfm.head(10)
rfm.to_csv("rfm.csv",index=False)
```

```python
# Segment-wise Revenue & Customer Count Summary (RFM segments  basis)


segment_summary_rfm = (
    rfm
    .groupby("Segment")
    .agg(
        customers_count=("customer_id", "count"),
        total_revenue=("total_revenue", "sum"),
        avg_revenue_per_customer=("total_revenue", "mean")
    )
    .reset_index()
    .sort_values("total_revenue", ascending=False)
    .round(2)
)

segment_summary_rfm
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")

# --- build clv_df ---
clv_df = (
    final_df.groupby("customer_id", as_index=False)
    .agg(
        total_revenue = ("line_revenue", "sum"),
        total_orders  = ("order_id", "nunique"),
        total_profit  = ("profit", "sum")
    )
)

# --- attach name (fix typo) ---
names = final_df.drop_duplicates("customer_id")[["customer_id","first_name"]]
clv_df = clv_df.merge(names, on="customer_id", how="left")
clv_df = clv_df.rename(columns={"first_name":"customer_name"})

# --- attach segment from rfm ---
clv_df = clv_df.merge(rfm[["customer_id","Segment"]], on="customer_id", how="left")

# --- realised & projected CLV ---
clv_df["CLV_realized"] = clv_df["total_profit"]

lifespan_map = {
    "Champions": 5,
    "Loyal Customers": 4,
    "Potential Loyalists": 3,
    "At Risk": 2,
    "Hibernating": 1,
    "Lost / Churn": 0
}
clv_df["expected_lifespan"] = clv_df["Segment"].map(lifespan_map).fillna(2)
clv_df["CLV_projected"] = clv_df["CLV_realized"] * clv_df["expected_lifespan"]

# --- QUICK CHECK: how many negatives before clipping ---
print("Negatives before clip - CLV_realized:", (clv_df["CLV_realized"] < 0).sum())
print("Negatives before clip - CLV_projected:", (clv_df["CLV_projected"] < 0).sum())

# --- CLIP negatives to 0 (remove log warnings & treat losses as 0 for CLV visual) ---
clv_df["CLV_realized"]  = clv_df["CLV_realized"].clip(lower=0)
clv_df["CLV_projected"] = clv_df["CLV_projected"].clip(lower=0)

# --- confirm after clipping ---
print("Negatives after clip - CLV_realized:", (clv_df["CLV_realized"] < 0).sum())
print("Negatives after clip - CLV_projected:", (clv_df["CLV_projected"] < 0).sum())

# --- Round for display ---
clv_df[["total_revenue","total_profit","CLV_realized","CLV_projected"]] = clv_df[["total_revenue","total_profit","CLV_realized","CLV_projected"]].round(2)

print(clv_df.head())
print(clv_df[["total_revenue","total_profit","CLV_projected"]].describe())
clv_df.to_csv("clv_projected_by_customer.csv", index=False)

```
